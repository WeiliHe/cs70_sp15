```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return (x-1)*(x-2)*(x-4)*(x-5)*(x-6) / (3-1)*(3-2)*(3-4)*(3-5)*(3-6)  

# Generate x values
x = np.linspace(-3, 3, 400)

# Generate y values
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = (x-1)*(x-2)*(x-4)*(x-5)*(x-6) / (3-1)*(3-2)*(3-4)*(3-5)*(3-6)')
plt.title('Plot of f(x) = (x-1)*(x-2)*(x-4)*(x-5)*(x-6) / (3-1)*(3-2)*(3-4)*(3-5)*(3-6)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
def extended_gcd(a, b):
    if b == 0:
        return (a, 1, 0)
    else:
        gcd, x1, y1 = extended_gcd(b, a % b)
        x = y1
        y = x1 - (a // b) * y1
        return (gcd, x, y)

# Example usage
a = 35
b = 11
gcd, x, y = extended_gcd(a, b)
print(f"GCD: {gcd}, x: {x}, y: {y}")
```
